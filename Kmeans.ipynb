```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
df.drop('CustomerID',axis=1,inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first=True, dtype='int')
```

```python
df
```

```python
sns.pairplot(df)
```

```python
km = KMeans(n_clusters=5)
```

```python
preds = km.fit_predict(df)
```
